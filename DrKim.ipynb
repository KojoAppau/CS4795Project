```python


from gensim.models import Word2Vec
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer
import nltk
from nltk.tokenize import word_tokenize

# Load pre-trained Word2Vec model (you would need to replace 'path_to_word2vec_model' with the actual path)
word2vec_model = Word2Vec.load('/Users/elijahappau/Downloads/GoogleNews-vectors-negative300.bin')

# Function to extract named entities from a sentence using Word2Vec
def extract_entities(sentence):
    tokens = word_tokenize(sentence)
    entities = [token for token in tokens if token in word2vec_model.wv.vocab]
    return entities

# Create a new chatbot instance
chatbot = ChatBot('NLPChatBot')

# Create a new trainer for the chatbot
trainer = ChatterBotCorpusTrainer(chatbot)

# Use the English language corpus data
trainer.train('chatterbot.corpus.english')

# Main chat loop
print("Bot: Hello! Ask me anything or type 'exit' to end the conversation.")

while True:
    # Get user input
    user_input = input("You: ")

    # Check for exit command
    if user_input.lower() == 'exit':
        print("Bot: Goodbye!")
        break

    # Extract named entities from user input using Word2Vec
    entities = extract_entities(user_input)

    # Print the extracted entities
    print(f"Entities: {entities}")

    # Get the bot's response
    response = chatbot.get_response(user_input)
    print(f"Bot: {response}")
```
