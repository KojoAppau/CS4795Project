```python
import spacy
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer

# Load spaCy English language model
nlp = spacy.load("en_core_web_sm")

# Function to extract named entities from a sentence
def extract_entities(sentence):
    doc = nlp(sentence)
    entities = [ent.text for ent in doc.ents]
    return entities

# Create a new chatbot instance
chatbot = ChatBot('NLPChatBot')

# Create a new trainer for the chatbot
trainer = ChatterBotCorpusTrainer(chatbot)

# Use the English language corpus data
trainer.train('chatterbot.corpus.english')

# Main chat loop
print("Bot: Hello! Ask me anything or type 'exit' to end the conversation.")

while True:
    # Get user input
    user_input = input("You: ")

    # Check for exit command
    if user_input.lower() == 'exit':
        print("Bot: Goodbye!")
        break

    # Extract named entities from user input
    
    entities = extract_entities(user_input)

    # Print the extracted entities
    print(f"Entities: {entities}")

    # Get the bot's response
    response = chatbot.get_response(user_input)
    print(f"Bot: {response}")

```

```python
import tkinter as tk
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer, ListTrainer

class ChatBotApp:
    def __init__(self, master):
        self.master = master
        master.title("ChatBot")

        self.chatbot = self.setup_chatbot()

        self.label = tk.Label(master, text="Ask me anything about science and math:")
        self.label.pack()

        self.user_input = tk.Entry(master, width=50)
        self.user_input.pack()

        self.chat_display = tk.Text(master, height=10, width=50)
        self.chat_display.pack()

        self.ask_button = tk.Button(master, text="Ask", command=self.ask_question)
        self.ask_button.pack()

    def setup_chatbot(self):
        chatbot = ChatBot("ScienceMathBot")

        # Use the English language corpus data for initial training
        trainer = ChatterBotCorpusTrainer(chatbot)
        trainer.train("chatterbot.corpus.english")

        # Additional training data for science and math
        trainer = ListTrainer(chatbot)
        trainer.train([
            "What is the Pythagorean theorem?",
            "The Pythagorean theorem states that in a right-angled triangle, the square of the hypotenuse is equal to the sum of the squares of the other two sides."
        
            "What does Elijah want today?",
            "No bending over and no akward answers"
            # Add more science and math questions and answers here
        ])

        return chatbot

    def ask_question(self):
        user_input = self.user_input.get()
        response = self.chatbot.get_response(user_input)

        # Display user input and chatbot response
        chat_display_text = self.chat_display.get("1.0", tk.END)
        self.chat_display.delete("1.0", tk.END)
        chat_display_text += f"You: {user_input}\nBot: {response}\n"
        self.chat_display.insert(tk.END, chat_display_text)

        # Train the chatbot with the new Q&A
        trainer = ListTrainer(self.chatbot)
        trainer.train([user_input, response])

        # Clear the user input field
        self.user_input.delete(0, tk.END)

if __name__ == "__main__":
    root = tk.Tk()
    app = ChatBotApp(root)
    root.mainloop()
```

Fine-tuning parameters in ChatterBot can be done by adjusting the configuration of the chatbot and trainer. Here are some key parameters you can consider modifying:

1. **Training Parameters:**
   - **`ChatterBotCorpusTrainer` parameters:**
     - `show_training_progress`: Set it to `False` to hide the training progress.
     - Example: `ChatterBotCorpusTrainer(chatbot, show_training_progress=False)`

   - **`ListTrainer` parameters:**
     - `show_training_progress`: Similar to above, controls whether training progress is displayed.
     - Example: `ListTrainer(chatbot, show_training_progress=False)`

2. **ChatBot Configuration:**
   - **`ChatBot` initialization parameters:**
     - `read_only`: Set it to `True` if you don't want the chatbot to learn from user interactions.
     - Example: `ChatBot("ExampleBot", read_only=True)`

   - **`ChatBot` method parameters:**
     - `get_response(statement, **kwargs)`: Adjust `threshold`, which controls the confidence level for response selection. A higher threshold makes responses more accurate but less diverse.
     - Example: `chatbot.get_response("What is the weather?", threshold=0.75)`

3. **Language Processing:**
   - **`ChatterBotCorpusTrainer` language parameter:**
     - `language`: Specify the language of the training data.
     - Example: `ChatterBotCorpusTrainer(chatbot, language='english')`

   - **`ListTrainer` language parameter:**
     - `language`: Similar to above, specify the language.
     - Example: `ListTrainer(chatbot, language='english')`

Adjusting these parameters allows you to control the training process, response confidence, and other aspects of the chatbot. Experiment with different settings and observe the impact on the bot's behavior. Keep in mind that fine-tuning may require some trial and error, and it's essential to evaluate the bot's performance on specific use cases.
